```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
train_df = pd.read_json('dataset/train.json')
```

```python
def calculate_passage_len(extracted_part):
    return extracted_part['answer_end'][0] - extracted_part['answer_start'][0]
```

```python
train_df['answer_length'] = train_df['extracted_part'].apply(lambda x: calculate_passage_len(x))
train_df['text_length'] = train_df['text'].apply(lambda x: len(x))
```

```python
train_df['ans_len/text_len'] = train_df['answer_length'] / train_df['text_length']
```
